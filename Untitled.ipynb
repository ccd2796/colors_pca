```python
from PIL import Image
import numpy as np

# Load the image
image = Image.open('grey_felt_hat.jpg')

# Convert the image to an RGB array
image_rgb = np.array(image)

print(image_rgb.shape)


import pandas as pd
import numpy as np

# Assuming image_rgb is your RGB numpy array from the previous step

# Flatten the image array and create a DataFrame
# Reshape from (height, width, 3) to (height*width, 3)
pixels = image_rgb.reshape(-1, 3)

# Create a DataFrame
df = pd.DataFrame(pixels, columns=['Red', 'Green', 'Blue'])

import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame with 'Red', 'Green', 'Blue' columns

# Plot the pairplot
sns.pairplot(df[['Red', 'Green', 'Blue']])

# Display the plot
plt.show()



```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming df is your DataFrame with 'Red', 'Green', 'Blue' columns

# Calculate the correlation matrix
corr_matrix = df[['Red', 'Green', 'Blue']].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .5})

plt.title('RGB Correlation Heatmap')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a figure and a 3D subplot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plotting
x = df['Red']
y = df['Green']
z = df['Blue']
ax.scatter(x, y, z, c=df[['Red', 'Green', 'Blue']]/255, marker='o')

# Labels
ax.set_xlabel('Red')
ax.set_ylabel('Green')
ax.set_zlabel('Blue')

# Adjust the view angle
ax.view_init(elev=20, azim=120)

# Optionally, adjust the margins of the plot or use tight_layout
plt.tight_layout()

# Display the plot
plt.show()
```
